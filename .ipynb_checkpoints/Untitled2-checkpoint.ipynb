```python
def plot_feature_importances(clf, feature_names):
    c_features = len(feature_names)
    plt.barh(range(c_features), clf.feature_importances_)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature name")
    plt.yticks(numpy.arange(c_features), feature_names)
```

## Log transformation

```python
num_df = df.select_dtypes('number')
num_df
```

```python
skew_val = df.skew(axis = 0)
skew_val
```

```python
field = 'purchase_price_1'
fig, (ax_before, ax_after) = plt.subplots(1, 2, figsize = (10, 5))
sns.distplot(df[field], ax = ax_before, fit = norm)
#plt.xlim(-100, 1750)
sns.distplot(df[field].apply(np.log1p), ax = ax_after, fit = norm)
ax_before.set(title='before np.log1p', ylabel='frequency', xlabel='value')
ax_after.set(title='after np.log1p', ylabel='frequency', xlabel='value')
fig.suptitle('Field "{}"'.format(field))
```

```python
for col in skew_val.index.values:
    if col == 'living_space_in_mÂ²':
        continue
    df[col] = df[col].apply(np.log1p)
```
