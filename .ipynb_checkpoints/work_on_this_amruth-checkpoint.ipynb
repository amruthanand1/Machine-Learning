### Here the feature is right-scewed so bringing it to normal distribution

```python
price_df = df[df['purchase_price'] < 1.21e+06]
fig, (ax_before, ax_after) = plt.subplots(1, 2, figsize = (10, 5))
sns.distplot(price_df['purchase_price'], fit = stats.norm, ax = ax_before)
sns.distplot(price_df['purchase_price'].apply(np.log1p), ax = ax_after, fit = stats.norm)
ax_before.set(title='before np.log1p', ylabel='frequency', xlabel='value')
ax_after.set(title='after np.log1p', ylabel='frequency', xlabel='value')
fig.suptitle('Field "{}"'.format('purchase_price'))
```
