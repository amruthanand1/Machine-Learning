```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_excel(r'F:\amruth\new_data.xlsx', encoding = 'iso-8859-1')
data.head()
```

```python
data.shape
```

```python
sns.heatmap(data.isnull())
```

```python
typo = data.type.unique().tolist()
#typo = typo.sort()
kind = sorted(typo)
kind
```

```python
from sklearn.preprocessing import OneHotEncoder
# creating one hot encoder object 
onehotencoder = OneHotEncoder()
#reshape the 1-D country array to 2-D as fit_transform expects 2-D and finally fit the object 
X = onehotencoder.fit_transform(data.type.values.reshape(-1,1)).toarray()
np.set_printoptions(threshold = np.inf)
X
```

```python
part_data = pd.DataFrame(X, columns = kind)
part_data
```

```python
df = pd.concat([data, part_data], axis = 1)
df
```

```python
#df[df['terrasse', 'balkon', 'garten', 'Aufzug'] == 'yes'].astype(float)
df['terrasse_1'] = (df.terrasse == 'yes').astype(float)
df['balkon_1'] = (df.balkon == 'yes').astype(float)
df['garten_1'] = (df.garten == 'yes').astype(float)
df['Aufzug_1'] = (df.Aufzug == 'yes').astype(float)
```

```python
df
```

```python
df.drop(['type', 'terrasse', 'balkon', 'garten', 'Aufzug', 'geo_point_2d','price_m2'], axis = 1, inplace = True)
df.shape
```
