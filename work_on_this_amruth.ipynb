# Outliers and modified boxplot

### Identifying outliers for modified boxplots
###### ===> Find Quartiles i.e. Q1(25%), Q2(50%), Q3(75%)
###### ===> Find InterQuartileRange(IQR) = Q3 - Q1
###### ===> Evaluate 1.5 * IQR

```python
#price_df = df['purchase_price'].apply(np.log1p)
Q1 = df['purchase_price'].quantile(0.25)
Q2 = df['purchase_price'].quantile(0.50)
Q3 = df['purchase_price'].quantile(0.75)
IQR = Q3 - Q1
by_an_amount = 1.5 * IQR
```

##### Any outliers are above Q3 by an amount 1.5*IQR or below Q1 by more than 1.5*IQR is an outlier.

```python
max_not_outlier = df['purchase_price'].quantile(0.75) + by_an_amount
min_not_outlier = df['purchase_price'].quantile(0.25) - by_an_amount
```

```python
fig, axes = plt.subplots(figsize = (10, 5))
sns.boxplot(df['purchase_price'], ax = axes, showfliers = True, showcaps = True)
plt.xlim(min_not_outlier, max_not_outlier)
```

#### Ohh, max_limit is in and around "1.21e6".
